```python
import numpy as np
import pandas as pd
import seaborn as sns
url='https://raw.githubusercontent.com/AdarSaban/olypmic_project/main/my_porject_csv.csv '
data=pd.read_csv(url)
data.Sport.unique()
data.Sport.value_counts(ascending=True)
```

```python
data["Age"] = data.groupby('Sport')['Age'].transform(lambda x: x.fillna(x.median()))
data["Height"] = data.groupby('Sport')['Height'].transform(lambda x: x.fillna(x.mean()))
data["Weight"] = data.groupby('Sport')['Weight'].transform(lambda x: x.fillna(x.mean()))
data
```
